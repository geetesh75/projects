```python
import os
def get_topic_page(topic_links):
    # Download the page
    response = requests.get(topic_links)
    # Check successful response
    if response.status_code !=200:
        raise Exception('Failed to load page{}'.format(topic_url))
    # parse using Beautiful soup
    topic_doc = BeautifulSoup(response.text,'html.parser')
    return topic_doc

def get_topic_info(ac_title_tag, ac_star_tag,ac_prices_tag ):
    # returns all the required info about a product
    product_name = ac_title_tag[0].text
    no_of_star = ac_star_tag[0].text
    prices = ac_prices_tag[0].text
    return product_name,no_of_star,prices

def get_topic_product(topic_doc):
    # Get the a tags containing product title
    ac_title_tag= topic_doc.find_all('h1', class_='yhB1nd')
    # Get star tags
    star_tag = topic_doc.find_all('div',class_='_3LWZlK')
    # Get price tag
    prices_tag = topic_doc.find_all('div',{'class':'_30jeq3 _16Jk6d'})
    
    topic_info_dict = {
        'product_title': [],
        'no_of_star':[],
         'prices':[] 
     } 
    # Get product info
    for i in range(len(ac_title_tag)):
        product_info = get_topic_info(ac_title_tag,star_tag,prices_tag) 
        topic_info_dict['product_title'].append(product_info[0])
        topic_info_dict['no_of_star'].append(product_info[1])
        topic_info_dict['prices'].append(product_info[2])
        
    return pd.DataFrame(topic_info_dict)

def scrape_topic(topic_links, path):
    if os.path.exists(path):
        print("The file {} already exists. skipping...".format(path))
        return
              
    df = get_topic_product(get_topic_page(topic_links))
    df.to_csv(path, index= None) 
```

```python
def get_topic_titles(doc):
    title_tags = doc.find_all('a', {'class':'s1Q9rs'})
    topic_titles = []
    for tag in title_tags:
        topic_titles.append(tag.text)
    return topic_titles

def get_topic_prices(doc):
    price_tags = doc.find_all('div', class_='_3I9_wc')
    prices = []
    for tag in price_tags:
        prices.append(tag.text)
    return prices
    

def get_topic_urls(doc):
    topic_link_tags = doc.find_all('a',{'class':'s1Q9rs'})
    topic_links = [] 
    base_url ="https://www.flipkart.com"  
    for tag in topic_link_tags:
        topic_links.append(base_url+ tag['href'])
    return topic_links
    
def scrape_topics():
    topics_url = 'https://www.flipkart.com/acnewclp-store'
    response = requests.get(topics_url,timeout=10)
    if response.status_code !=200:
        raise Exception('Failed to load page{}'.format(topic_url))
    topics_dict = {'title':get_topic_titles(doc),'prices':get_topic_prices(doc),'url':get_topic_urls(doc)}
    
    return pd.DataFrame( topics_dict)
```

```python
def scrape_topic_product():
    print('scraping list of product')
    df = scrape_topics()
    # create folder here
    os.makedirs('data', exist_ok=True)
    
    for index, row in df.iterrows():
        print('scraping topic product for "{}"'.format(row['title']))
        scrape_topic(row['url'],'data/{}.csv'.format(row['title'])) 
    
```

```python
def get_page_topic (page_number):
    url = 'https://www.flipkart.com/acnewclp-store?page='+ str(page_number)
    doc = get_topic_page(url)
    title = get_topic_titles(doc)
    prices = get_topic_prices(doc)
    discount = get_discount_prices(doc)
    urls = get_topic_urls(doc)
    return title,prices,discount,urls
```
